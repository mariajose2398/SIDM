```python
def sum_bgs_hist(bgs_list, histogram_name, channel_name, dir_name):
    """
    Sum histograms for a list of backgrounds.
    """
    summed_hist = None
    
    for x in bgs_list:
        # print(x)
        output_bg = coffea.util.load(f"{dir_name}/output_{x}.coffea")
        # print(output_bg["out"][x]["hists"].keys())

        hist = output_bg["out"][x]["hists"][histogram_name][channel_name, :50j]
        
        if summed_hist is None:
            summed_hist = hist.copy()
        else:
            summed_hist += hist
    
    return summed_hist
channel_name = "base_ljObjCut_ljIso_2lj"
hist_collections = "displacement_base"
dir_name = channel_name+hist_collections
output_signal = coffea.util.load(f"{dir_name}/output_signal.coffea" )
# print(output_signal["out"]["2Mu2E_500GeV_0p25GeV_0p04mm"]["hists"].keys())

histogram_list  = [ 
 "mu_lj_pfMuon_dxy_signi",
# "mu_lj_dsaMuon_dxy_signi",
"egm_lj_electron_dxy_signi",
 "mu_lj_pfMuon_min_dxy_signi",
"mu_lj_pfMuon_max_dxy_signi",
# "mu_lj_dsaMuon_min_dxy_signi",
# "mu_lj_dsaMuon_max_dxy_signi",
 "egm_lj_electron_min_dxy_signi",
 "egm_lj_electron_max_dxy_signi",
    "mu_lj_pfMuon_dxyErr",
# "mu_lj_dsaMuon_dxyErr",
"egm_lj_electron_dxyErr",
"leading_mu_lj_pf_muon_min_dxy_signi",
"leading_mu_lj_pf_muon_max_dxy_signi", 
# "leading_mu_lj_pf_muon_min_dxy", 
# "leading_egm_lj_electron_min_dxy", 
"leading_egm_lj_electron_min_dxy_signi"

    

]
S = [
   
'2Mu2E_500GeV_0p25GeV_0p04mm',
'2Mu2E_500GeV_0p25GeV_2p0mm',
'2Mu2E_500GeV_5p0GeV_0p8mm',
'2Mu2E_500GeV_5p0GeV_40p0mm',

]
# sum_bg = sum_bgs_hist(QCD, histogram_name,channel_name )
for i, histogram_name in enumerate(histogram_list):
    # print(i, histogram_name, title_list[i])
    # plt.figure(figsize=(15, 12))
    sum_bg_qcd = sum_bgs_hist(QCD, histogram_name,channel_name, dir_name )
    sum_bg_tt = sum_bgs_hist(TT, histogram_name,channel_name, dir_name )
    sum_bg_dy = sum_bgs_hist(DY, histogram_name,channel_name, dir_name )
    
    utilities.plot(
    [
        sum_bg_qcd, 
     sum_bg_tt,
      sum_bg_dy
    ],
    flow="none",
    yerr=True,
    density=True,
    histtype="fill",
    color=["#A9A9A9",
           "#FFB6C1",
           "#9ECAE1"
          ],
    label=[
        "QCD", 
           "TT",
         r"DY$\rightarrow\mu\mu$"
          ],
    stack=True,
)
    labels = ["3cm", "150cm", "3cm", "150cm"]
    colors = ['r', 'g', 'b', 'purple']
    for i, s in enumerate(S):
        print(s)
        parts = s.split("_")
        mass = parts[1]
        zd_mass = parts[2].replace("p", ".")
        ctau = parts[3].replace("p", ".").replace("mm", "")
        label =  zd_mass + ", " + labels[i]
        histogram = output_signal["out"][s]["hists"][histogram_name][channel_name, :50j]

    # print(histogram.values())
        utilities.plot(histogram, flow ="none", yerr=True, density=True, label=label, color=colors[i], linewidth=3)
    hep.label.exp_text(text="500 GeV ", loc=3) 
    plt.yscale("log")
    # plt.title(histogram_name)
    plt.legend()
    plt.ylim(1e-3, 1e0)
    plt.savefig("/home/cms-jovyan/SIDM/plots/" + histogram_name + ".png", dpi=300)
    plt.show()
    # plt.savefig("/home/cms-jovyan/SIDM/plots/" + histogram_name + ".png")
    plt.close()
```
