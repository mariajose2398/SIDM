```python
client = scaleout.make_dask_client("tls://localhost:8786")
client
```

```python
runner = processor.Runner(
    # executor=processor.IterativeExecutor(),
    executor=processor.DaskExecutor(client=client),
    # schema=NanoAODSchema,
    schema = llpnanoaodschema.LLPNanoAODSchema,
    #maxchunks=1,
    skipbadfiles=True,
)
channels = [
    # "baseNoLj", 
    "baseNoLj_A_mumu", 
    "baseNoLj_A_mumu_matched_lj",
    "baseNoLj_A_mumu_matched_lj_iso",
    "baseNoLj_A_mumu_matched_lj_iso_disp"
    
    
        
           ]

p = sidm_processor.SidmProcessor(
    channels,
    ["lj_reco_effi_base", ],
)
```

```python
 samples = [

     '2Mu2E_500GeV_0p25GeV_4p0mm',
     '2Mu2E_500GeV_1p2GeV_19p0mm',
     '2Mu2E_500GeV_5p0GeV_80p0mm',
     

    
]
fileset = utilities.make_fileset(samples, "llpNanoAOD_v2", max_files=-1, location_cfg="signal_2mu2e_v10.yaml")

                            

output_signal = runner.run(fileset, treename="Events", processor_instance=p)
coffea.util.save(output_signal, "output_files/output_signal_effi_muons_lxy.coffea"  )
```

```python

hists = output_signal["out"]["2Mu2E_500GeV_0p25GeV_4p0mm"]["hists"]
ch1 = "baseNoLj_A_mumu"
ch2 = "baseNoLj_A_mumu_matched_lj"
ch3 = "baseNoLj_A_mumu_matched_lj_iso"
ch4 = "baseNoLj_A_mumu_matched_lj_iso_disp"

legend = [
r"$Z_d \rightarrow \mu\mu $",
r"$Z_d \rightarrow \mu\mu$ with $\mu$ LJ",
r"$Z_d \rightarrow \mu\mu$ with $\mu$ LJ with Iso",
r"$Z_d \rightarrow \mu\mu$ with $\mu$ LJ with Iso and Disp"
]
histogram = "genAs_toMu_pt_highRange"

num_list = [
          hists[histogram][ch2, :400j],
          hists[histogram][ch3, :400j],
         hists[histogram][ch4, :400j],
]
den = hists[histogram][ch1,:400j]

utilities.plot_ratio (num_list,den,legend =legend, text = "500GeV, 0.25GeV, 300 cm", ylabel = "Arbitrary Units")
folder_name = "plots/"
file_name = "2Mu2E_500GeV_0p25GeV_4p0mmgenAs_toMu_pt"
save_path = os.path.join(folder_name, file_name)         
    
plt.savefig(save_path)

    
```
